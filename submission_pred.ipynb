```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Daten laden
df = pd.read_csv('submission.csv')
print(df.columns)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
# split the data into training and test sets
df_train, df_test = train_test_split(df, test_size=0.3, random_state=42)

# Daten vorbereiten
X = df_train[['Speed', 'Load']]  # Features
y = []
y.append(df_train['Railpressure'])  
y.append(df_train['Air'])
y.append(df_train['Angle'])
y.append(df_train['Pressure1'])
y.append(df_train['Pressure2'])
y.append(df_train['Temperature'])

X_trains, X_tests, y_trains, y_tests = [], [], [], []
for i in y:
    X_train, X_test, y_train, y_test = train_test_split(X, i, test_size=0.2, random_state=42)
    X_trains.append(X_train)
    X_tests.append(X_test)
    y_trains.append(y_train)
    y_tests.append(y_test)

models = []
y_preds = []
mses = []
print('Training:')
for i in range(len(y)):
    # Modell ausw√§hlen und trainieren
    model = LinearRegression()
    #model = RandomForestRegressor(n_estimators=100, random_state=42)
    #model = DecisionTreeRegressor(random_state=42)
    #model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
    model.fit(X_trains[i], y_trains[i])
    models.append(model)

    # Vorhersagen treffen
    y_pred = model.predict(X_tests[i])
    y_preds.append(y_pred)

    # Modell bewerten
    mse = mean_squared_error(y_tests[i], y_pred)
    mses.append(mse)
    print(f'Mean Squared Error: {mse}')

print('Validation:')

# Validate the model by predicting the test data
X_test = df_test[['Speed', 'Load']]
y_test = []
y_test.append(df_test['Railpressure'])
y_test.append(df_test['Air'])
y_test.append(df_test['Angle'])
y_test.append(df_test['Pressure1'])
y_test.append(df_test['Pressure2'])
y_test.append(df_test['Temperature'])

for i in range(len(y)):
    y_pred = models[i].predict(X_test)
    mse = mean_squared_error(y_test[i], y_pred)
    print(f'Mean Squared Error: {mse}')
    # print(f'Prediction: {y_pred}')
    # print(f'Actual: {y_test[i]}')
    # print()

# show of mse is big in comparison to the ranges of the values
print('Ranges:')
print(df_train.describe())
print()
print('MSEs:')
print(mses)
```
